```python
#set union and interection
```

```python
a = {10,20,30}
b = {10,100,200}

u = a.union(b)
print(u)
```

```python
i = a.intersection(b)
print(i)
```

```python
#subset,superset,disjoint
```

```python
a = {10,20,30}
b = {10,20,30,40,50,60,70}

print(a.issubset(b))
print(b.issubset(a))
```

```python
print(a.issuperset(b))
print(b.issuperset(a))
```

```python
#disjoint - no common items
print(a.isdisjoint(b))
```

```python
#set differemnce and symmetric difference
```

```python
a = {10,20,30}
b = {10,20,100}
```

```python
print(a.difference(b))
```

```python
print(a.symmetric_difference(b))
```
