```python
#while condition:
#    statements
```

```python
num = int(input("Enter a number"))

count = 0

while num > 1:
    num >> 1
    count = count +1

print(count)


```

```python
i = 5

while i >= 1:
    print(i)
    i = i-1
```

```python
#NESTED LOOP
```

```python
i = 1
while i <= 3:
    print("Outer Loop Iteration",i)
    j=i
    while j <= 3:
        print("Inner Loop Interation",j)
        j += 1
    i += 1
    print("===========================")
```

```python
list = [10,20,30]

for x in list:
    print(x)
else:
    print("I will be executed")
```
