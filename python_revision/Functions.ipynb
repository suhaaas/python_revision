```python
#factorial using recursion
```

```python
def factorial(n):
    if n<=1:
        return n
    return n * factorial(n-1)

n = int(input())

if n>=0:
    ans = factorial(n)
    print(ans)
```

```python
#count vowels
```

```python
def count_vowels(s):
    vowels = "aeiou"
    return sum(ch in vowels for ch in s)

print(count_vowels("python functions"))
```

```python
text = "python functions"

def count_vowels(s):
    vowels = "aeiou"
    count = 0
    for ch in s:
        if ch in vowels:
            count +=1
    return count

print(count_vowels(text))
```

```python
#reverse a number

num = 1234

def reverse_number(n):
    rev = 0
    while n > 0:
        digit = n % 10
        rev = rev*10+digit
        n//=10
    return rev

print(reverse_number(num))
```

```python
#find factorial
```

```python
num = 6

def factorial(n):
    result = 1
    while n > 0:
        result *= n
        n -= 1
    return result

print(factorial(num))
```
