```python
#printing second largest numbers
numbers = [10,20,60,90,30,40,50]
numbers.sort()

print(numbers)
print('second largest number',numbers[-2])
```

```python
#return only even numbers
numbers = [10,20,30,33,1,11,7,40,50]

even = []

for n in numbers:
    if n % 2 == 0:
        even.append(n)

print("even numbers are", even)
```
