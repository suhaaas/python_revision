```python
#DICTIONARIES
```

```python
data = {"Name":"Pyhton","Pages":245,"Price":100,"Ratings":6}
```

```python
data
```

```python
#accessing dict items
data.get("Name")
```

```python
#check if exists
var = "Pages"

if var in data:
    print(data[var])
else:
    print("not found")
```

```python
#access dic using for loop
data = {"Name":"Pyhton","Pages":245,"Price":100,"Ratings":6}
for x in data:
    print(x)
```

```python
for x in data:
    print(data[x])
```

```python
for x in data.keys():
    print(x)
```

```python
for x,y in data.items():
    print(x,y)
```

```python
#modifictions in dic
data = {"Name":"Pyhton","Pages":245,"Price":100,"Ratings":6}
data['Price'] = 750
```

```python
data
```

```python
data['Author'] = "rcb"
print(data)
```

```python
#remove item
page_count = data.pop("Author")
print(page_count)
```
